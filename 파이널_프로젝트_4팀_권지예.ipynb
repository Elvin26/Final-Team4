## 어떤 카테고리가 가장 많이 팔렸을까?

### 소분류

```python
# '상품카테고리'별 판매량 집계
category_counts = df2['상품카테고리'].value_counts()

# 전체 판매량 계산
total_sales = category_counts.sum()

# 각 카테고리별 판매 비중 계산
category_proportions = category_counts / total_sales

# 결과를 데이터프레임으로 통합
category_analysis = pd.DataFrame({
    '판매량': category_counts,
    '비중': category_proportions
})

# 판매량 기준으로 내림차순 정렬하여 결과 출력
display(category_analysis.sort_values(by='판매량', ascending=False))
```

```python
import matplotlib.cm as cm
import numpy as np
import matplotlib.colors as mcolors

# 상위 20개 카테고리 선택
top_20_categories = category_analysis.sort_values(by='판매량', ascending=False).head(20)

# 누적 비중 계산
top_20_categories['누적 비중'] = top_20_categories['비중'].cumsum()

# 막대 그래프 및 라인 그래프를 위한 figure 및 axes 생성
fig, ax1 = plt.subplots(figsize=(12, 6))

# 막대 색상 설정을 위한 사용자 지정 컬러맵 생성 (빨간색 - 분홍색)
colors = ["yellowgreen", "indianred"] # from lower values (pink) to higher values (red)
cmap = mcolors.LinearSegmentedColormap.from_list("red_green_gradient", colors)


# 막대 그래프 (판매량)
bars = ax1.bar(top_20_categories.index, top_20_categories['판매량'], color=cmap(np.linspace(1, 0, len(top_20_categories))))
ax1.set_title('상위 20개 상품 카테고리별 판매량 및 누적 비중')
ax1.set_xlabel('상품 카테고리')
ax1.set_ylabel('판매량')
ax1.tick_params(axis='y')
# Explicitly set tick locations before setting labels to address the warning
ax1.set_xticks(range(len(top_20_categories.index)))
ax1.set_xticklabels(top_20_categories.index, rotation=45, ha='right')

# 막대 위에 판매량 텍스트 표시
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 0), va='bottom', ha='center') # 판매량은 정수로 표시

# 라인 그래프 (누적 비중)를 위한 두 번째 y축 생성
ax2 = ax1.twinx()
line, = ax2.plot(top_20_categories.index, top_20_categories['누적 비중'], color='red', linestyle='-') # 누적 비중 라인 색상 변경
ax2.set_ylabel('전체 판매량 중 누적 비중', color='red') # 누적 비중 라벨 색상 변경
ax2.tick_params(axis='y', labelcolor='red') # 누적 비중 y축 틱 색상 변경

# 라인 위에 누적 비중 텍스트 표시
for x, y in zip(top_20_categories.index, top_20_categories['누적 비중']):
    ax2.text(x, y, f'{y:.2f}', va='bottom', ha='center', color='black') # 누적 비중은 소수점 둘째 자리까지 표시, 텍스트 색상 변경

# 레이아웃 조정 및 그래프 표시
fig.tight_layout()
plt.show()
```

#### 1. 상위 판매자들의 특징:

- 이들 상위 판매자들 간에도 총 판매 건수에는 상당한 차이가 존재하며, 이는 판매 전략이나 전문 분야에 따라 성과가 다를 수 있음을 시사합니다.

#### 2. 판매 지역과의 관계:

- 대부분의 상위 판매자들은 브라질의 주요 경제 중심지인 SP(상파울루) 주에서 가장 높은 판매량을 기록했습니다.

- 일부 상위 판매자는 SP 외의 특정 주에서도 높은 판매량을 보이며, 이는 해당 지역 시장에서의 강점을 가지고 있거나 특정 지역 고객층을 타겟으로 하는 전략이 유효함을 보여줍니다.
- 반면, 일부 판매자는 여러 주에 걸쳐 비교적 고르게 판매 활동을 하여 지리적 분산을 통한 안정적인 판매를 추구하는 것으로 보입니다.

#### 3. 상품 카테고리와의 관계:

- 상위 판매자들은 주로 특정 상품 카테고리에서 높은 판매량을 보였습니다. 특히 '침구/욕실/테이블', '건강/뷰티', '스포츠/레저', '가구/인테리어', '컴퓨터 액세서리' 등 인기 카테고리에서 강세를 나타냈습니다.

- 일부 판매자는 소수의 특정 카테고리에 집중하여 전문성을 강화하는 전략을 사용하는 것으로 보이며, 다른 판매자는 여러 카테고리를 다루며 다양한 고객층을 확보하려는 전략을 사용하는 것으로 보입니다.

#### 4. 인사이트 및 결론:

- 판매 성과는 판매자의 지리적 도달 범위와 주력 상품 카테고리에 밀접하게 연결되어 있습니다.

- 인기 상품 카테고리에 집중하거나 특정 카테고리에서 전문성을 확보하는 것이 판매 경쟁력을 높이는 데 유리할 수 있습니다.
- 하지만 특정 지역 또는 틈새 카테고리에 집중하는 전략도 성공적인 판매 성과를 이끌어낼 수 있습니다. 중요한 것은 판매자의 강점과 시장의 특성을 고려한 적절한 지역 및 카테고리 선택입니다.

## 판매자 ID별 조사 판매량 기준 분석

```python
# 판매자 ID별 총 판매 금액 집계
seller_revenue = df2.groupby('판매자_ID')['결제금액'].sum().reset_index(name='총판매금액')

# 판매자 주 정보를 가져오기 위해 df2와 병합
# 판매자 ID별 고유한 판매자 주를 가져오기 위해 drop_duplicates 사용
seller_state = df2[['판매자_ID', '판매자_주']].drop_duplicates()
seller_revenue = pd.merge(seller_revenue, seller_state, on='판매자_ID', how='left')

# '판매자_ID' 컬럼을 '판매자_ID(판매자_주)' 형식으로 변경
seller_revenue['판매자_ID'] = seller_revenue['판매자_ID'].astype(str) + '(' + seller_revenue['판매자_주'].astype(str) + ')'

# 더 이상 필요 없는 '판매자_주' 컬럼 삭제
seller_revenue = seller_revenue.drop('판매자_주', axis=1)

# 총 판매 금액 기준으로 내림차순 정렬
seller_revenue_sorted = seller_revenue.sort_values(by='총판매금액', ascending=False)

display(seller_revenue_sorted.head())
```

```python
import matplotlib.cm as cm
import numpy as np
import matplotlib.colors as mcolors

# 판매 금액 기준으로 판매자 ID 0번 제외
filtered_seller_revenue = seller_revenue_sorted[seller_revenue_sorted['판매자_ID'] != '0(nan)'].copy()


# 판매 금액 기준으로 상위 15개 판매자 추리기 (copy()를 사용하여 경고 방지)
top_sellers_revenue = filtered_seller_revenue.head(15).copy()

# 누적 총 판매 금액 계산
top_sellers_revenue['누적 총판매금액'] = top_sellers_revenue['총판매금액'].cumsum()

# 전체 판매 금액 계산
total_revenue_all_sellers = filtered_seller_revenue['총판매금액'].sum() # 전체 판매 금액 계산 시에도 0번 판매자 제외

# 누적 총 판매 금액의 전체 판매량 대비 비중 계산
top_sellers_revenue['누적 비중'] = top_sellers_revenue['누적 총판매금액'] / total_revenue_all_sellers


# Create a bar plot for total revenue of top sellers
fig, ax1 = plt.subplots(figsize=(14, 7))

# 막대 색상 설정을 위한 사용자 지정 컬러맵 생성 (yellowgreen - indianred)
colors = ["yellowgreen", "indianred"]
cmap = mcolors.LinearSegmentedColormap.from_list("yellowgreen_indianred_gradient", colors)

# 막대 그래프 (총 판매 금액)
bars = ax1.bar(top_sellers_revenue['판매자_ID'].astype(str), top_sellers_revenue['총판매금액'], color=cmap(np.linspace(1, 0, len(top_sellers_revenue))))

ax1.set_title('판매자 상위 15명의 총 판매 금액 및 누적 비중')
ax1.set_xlabel('판매자 ID(판매자 주)')
ax1.set_ylabel('총 판매 금액')
ax1.tick_params(axis='y')
# Explicitly set tick locations before setting labels to address the warning
ax1.set_xticks(range(len(top_sellers_revenue['판매자_ID'])))
ax1.set_xticklabels(top_sellers_revenue['판매자_ID'].astype(str), rotation=45, ha='right')

# 라인 그래프 (누적 비중)를 위한 두 번째 y축 생성
ax2 = ax1.twinx()
ax2.plot(top_sellers_revenue['판매자_ID'].astype(str), top_sellers_revenue['누적 비중'], color='red', marker='o', linestyle='-')
ax2.set_ylabel('전체 판매 금액 중 누적 비중', color='red')
ax2.tick_params(axis='y', labelcolor='red')
ax2.grid(False) # 보조축의 그리드 숨김

# 막대 위에 총 판매 금액 텍스트 표시
for bar in bars:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.0f}', va='bottom', ha='center') # 총 판매 금액은 정수로 표시

# 라인 위에 누적 비중 텍스트 표시
for x, y in zip(top_sellers_revenue['판매자_ID'].astype(str), top_sellers_revenue['누적 비중']):
    ax2.text(x, y, f'{y:.2f}', va='bottom', ha='center', color='black') # 누적 비중은 소수점 둘째 자리까지 표시


plt.tight_layout()
plt.show()
```
